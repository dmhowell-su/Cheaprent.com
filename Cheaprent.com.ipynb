```python
#mark on map where different locations are
import folium #street view, great zoom
Rental_Location_Syracuse = (43.088947,-76.154480) #autozoom on Syracuse, NY
map = folium.Map(location=Rental_Location_Syracuse, zoom_start=11) #how close the zoom is
map

for row in combined.to_records(): 
    pos = (row['Lat'],row['Long']) #lat and long from DataFrame
    message = "%s in %s,%s" % (row['Company_y'],row['City_y'], row['State_y']) #column name from Rental Address DataFrame
    marker = folium.Marker(location=pos, 
                    popup=message
                          )
    map.add_children(marker)
map
```
